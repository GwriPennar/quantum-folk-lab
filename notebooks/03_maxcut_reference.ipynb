# EXP-002 Max-Cut Reference

This notebook is a thin companion to the package implementation. It uses the registered `cycle4` graph to show exact enumeration, the QUBO/Ising convention, and the local Qiskit QAOA path.

Max-Cut assigns each graph node to one of two partitions. An edge contributes its weight when its endpoints are in different partitions. For `cycle4`, the complementary optimal assignments are `0101` and `1010` in graph-node order.

```python
from quantum_folk_lab.maxcut import get_graph
from quantum_folk_lab.maxcut_exact import solve_maxcut_exact

graph = get_graph("cycle4")
exact = solve_maxcut_exact(graph)
exact.to_dict()
```

The binary objective uses `different(i, j) = x_i + x_j - 2 x_i x_j`. The Pauli cut operator is `H_cut = sum w_ij / 2 * (I - Z_i Z_j)`. The QAOA optimiser minimises `-H_cut`, so expected cut is `-expected_energy`.

```python
from quantum_folk_lab.maxcut_ising import verify_equivalence

rows = verify_equivalence(graph)
[row.to_dict() for row in rows]
```

Qiskit bit ordering is handled by package conversion functions. Qubit `k` represents graph node `nodes[k]`; sampler count keys are decoded back to graph-node order before scoring.

```python
from quantum_folk_lab.maxcut_qaoa import run_maxcut_qaoa

qaoa = run_maxcut_qaoa(graph, depth=1, shots=4096, seed=42)
qaoa.to_dict()
```

The sampled best bitstring can be optimal even when the expected approximation ratio is below 1. This is an ideal local simulation and not a hardware run.
